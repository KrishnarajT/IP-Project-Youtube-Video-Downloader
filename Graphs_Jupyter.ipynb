```python
# here is an example graph
import numpy as np
import File_IO as fio
import matplotlib.pyplot as plt

video_views = fio.read.get_views()
views_list = []
for views in video_views:
    views_list.append(int(views.rstrip('\n')))
plt.xlabel('Videos')
plt.ylabel('views')
print(views_list)
plt.bar(np.arange(24), views_list)
plt.show()
```
